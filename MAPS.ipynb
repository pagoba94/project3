```python
df_sanfrancisco = pd.read_csv("dataframes/df_sanfrancisco.csv")
df_sanfrancisco
```

```python
import folium
import pandas as pd

# we create a map centered on San Francisco coordinates
sanfrancisco_map = folium.Map(location=[37.7749295, -122.4194155], zoom_start=12)

# Iterate over the rows of the DataFrame
for index, row in df_sanfrancisco.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Add a marker to the map for each office
    folium.Marker([latitude, longitude], popup=name).add_to(sanfrancisco_map)
    
#show the map
sanfrancisco_map
```
