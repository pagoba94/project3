```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
df_sanfrancisco = pd.read_csv("dataframes/df_sanfrancisco.csv")
df_sanfrancisco
```

```python
sf_lat= "37.7968917"
sf_lon= "-122.4032027"

# we create a map centered on San Francisco coordinates
sanfrancisco_map = folium.Map(location=[sf_lat, sf_lon], zoom_start=10)

design_companies = df_sanfrancisco[df_sanfrancisco['data'] == 'Design company']
# Iterate over the rows of the DataFrame
for index, row in design_companies.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='pink', icon='briefcase', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)
    
data = df_sanfrancisco[df_sanfrancisco['data'] == 'Airport']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='blue', icon='plane', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)


data = df_sanfrancisco[df_sanfrancisco['data'] == 'Preschool']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='lightblue', icon='baby-carriage', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)

data = df_sanfrancisco[df_sanfrancisco['data'] == 'Train Station']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='darkred', icon='train', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)

data = df_sanfrancisco[df_sanfrancisco['data'] == 'Night Club']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='darkpurple', icon='martini-glass', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)

data = df_sanfrancisco[df_sanfrancisco['data'] == 'Vegan Restaurant']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='green', icon='utensils', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)

data = df_sanfrancisco[df_sanfrancisco['data'] == 'Basketball stadium']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='orange', icon='basketball', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)

data = df_sanfrancisco[df_sanfrancisco['data'] == 'starbucks']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='cadetblue', icon='mug-saucer', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)

data = df_sanfrancisco[df_sanfrancisco['data'] == 'dog grooming']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='gray', icon='dog', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)

data = df_sanfrancisco[df_sanfrancisco['data'] == 'Primary and Secondary School']
# Iterate over the rows of the DataFrame
for index, row in data.iterrows():
    name = row['name']
    latitude = row['latitude']
    longitude = row['longitude']
    
     # Create a custom icon for the design company
    icon = folium.Icon(color='darkblue', icon='graduation-cap', prefix='fa')
    
    # Add a marker to the map for each office
    folium.Marker([latitude, longitude], icon=icon, popup=name).add_to(sanfrancisco_map)
    

icon = Icon (color = "red", icon_color = "white", icon = "star", prefix = "fa")
sf_location = Marker(location = [sf_lat, sf_lon], tooltip = "San Francisco Location", icon = icon)
sf_location.add_to(sanfrancisco_map)

sanfrancisco_map
```
